```python
data=[fifa]
for dataset in data:
    dataset['team']=dataset['Team'].map(country_mapping)
```

```python
for dataset in data:
    dataset['opponent']=dataset['Opponent'].map(country_mapping)
```

```python
fifa.head()
```

```python
binary_mapping = {'Yes':1,'No':0}
```

```python
for dataset in data:
    dataset['Binary_MOM']=dataset['Man of the Match'].map(binary_mapping)
```

# Data_Visualization

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sb.pairplot(fifa_new.drop(['Man of the Match'], axis=1), hue='Binary_MOM')
```

```python
def bar_chart(feature):
    MOM=fifa_new[fifa_new['Binary_MOM']==1][feature].value_counts()
    non_MOM=fifa_new[fifa_new['Binary_MOM']==0][feature].value_counts()
    df=pd.DataFrame([MOM, non_MOM])
    df.index=['Man of the match','Not man of the match']
    df.plot(kind='bar', stacked=True, figsize=(10,5))
```

```python
bar_chart('Pass Accuracy %')
```

# Modelling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score
```

```python
k_fold = KFold(n_splits=40, shuffle=True, random_state=0)
```

```python
train=model_df.drop(['Binary_MOM'],axis=1)
```

```python
target=model_df.loc[:,['Binary_MOM']]
```

```python
clf=RandomForestClassifier(n_estimators=100)
```

```python
score_2 = cross_val_score(clf, train, target, cv=k_fold, n_jobs=1)
print(score_2)
```

```python
round(np.mean(score_2)*100,2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf_2=GaussianNB()
```

```python
score=cross_val_score(clf_2, train, target, cv=k_fold, n_jobs=4)
print(score)
round(np.mean(score)*100,2)
```
